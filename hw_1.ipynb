```
import numpy as np
import matplotlib.pyplot as plt
```

```
# обратная функция
def inverse(y):
    return 1 - y
```

```
# пересечение
def max_min_intersection(y1, y2):
    # intersection - min
    if y1 < y2:
        return y1
    return y2
```

```
# обьединение
def max_min_union(y1, y2):
    # union - max
    if y1 > y2:
        return y1
    return y2
```

```
# пересечение
def algebra_intersection(y1, y2):
    return y1 * y2
```

```
# обьединение
def algebra_union(y1, y2):
    return y1 + y2 - y1 * y2
```

```
# пересечение
def limited_intersection(y1, y2):
    if y1 + y2 - 1 < 0:
        return 0
    return y1 + y2 - 1
```

```
# обьединение
def limited_union(y1, y2):
    if y1 + y2 > 1:
        return 1
    return y1 + y2
```

```
# делает функцию прямой по двум координатам
def get_line_by_two_point(a: tuple, b: tuple):
    def func(x):
        return (x - a[0]) * (b[1] - a[1]) / (b[0] - a[0]) + a[1]
    return func
```

```
def func_A(x):
    if x <= 3:
        return 0
    if x > 3 and x <= 5:
        f = get_line_by_two_point((3, 0), (5, 1))
        return f(x)
    if x > 5:
        return 1
```

```
def func_B(x):
    if x >= 0 and x <= 3:
        f = get_line_by_two_point((0, 0), (3, 1))
        return f(x)
    if x > 3 and x <= 5:
        f = get_line_by_two_point((3, 1), (5, 0))
        return f(x)
    if x > 5:
        return 0
```

```
def func_C(x):
    if x >= 0 and x <= 5:
        f = get_line_by_two_point((0, 1), (5, 0))
        return f(x)
    if x > 5:
        return 0
```

```
# функция отрисовки графика функции
# для функций A B C
def plot_function(func):
    dx = 0.01
    x = np.linspace(0, 16, 1000)
    y = np.array([func(x_i) for x_i in x])

    plt.xlim(0, 16)
    plt.ylim(0, 1.5)

    plt.grid(True)
    plt.plot(x, y)
```

```
plot_function(func_A)
```

```
plot_function(func_B)
```

```
plot_function(func_C)
```

```
# результирующая функция
def func_algebra_D(x):
    return algebra_union(algebra_intersection(func_A(x), func_B(x)), inverse(func_C(x)))
```

```
plot_function(func_algebra_D)
```

```
# результирующая функция
def func_max_min_D(x):
    return max_min_union(max_min_intersection(func_A(x), func_B(x)), inverse(func_C(x)))
```

```
plot_function(func_max_min_D)
```

```
# результирующая функция
def func_limited_D(x):
    return limited_union(limited_intersection(func_A(x), func_B(x)), inverse(func_C(x)))
```

```
plot_function(func_limited_D)
```

```
def small_price(x):
  if x >= 1 and x < 3:
    return 1
  if x >= 3 and x <= 5:
    f = get_line_by_two_point((3, 1), (5, 0)) 
    return f(x)
  return 0
```

объем закупок - <мало> = 0.33333333333333337

объем закупок - <средне> = 1.0

объем закупок - <много> = 0

```
print(small_quality(0.55))
print(medium_quality(0.55))
print(many_quality(0.55))
```

качество - <мало> = 0.24999999999999967

качество - <средне> = 0.3333333333333336

качество - <много> = 0

Правило 1: min (0.33333333333333337, max(0.24999999999999967, 0.3333333333333336)) = 0.33333333333333337;

Правило 2: min (0.33333333333333337, 0) = 0;

Правило 3: min (max(0.33333333333333337, 1.0), 0.3333333333333336) = 0.3333333333333336;

Правило 4: min (0, 0) = 0

```
def res_graph(arr, probability):
  res_arr = []
  for y in arr:
    if y <= probability:
      res_arr.append(y)
    else:
      res_arr.append(probability)
      
  return res_arr
```

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

axis_value = [0, 11, -0.5, 1.5]
plt.figure(figsize=(20, 4))
arr_x = np.linspace(0, 10, 1000)

func_1 = res_graph([small_price(x) for x in arr_x], 0.33333333333333337)
func_2 = res_graph([medium_price(x) for x in arr_x], 0)
func_3 = res_graph([medium_price(x) for x in arr_x], 0.3333333333333336)
func_4 = res_graph([many_price(x) for x in arr_x], 0)

# Правило 1 - мало
plt.subplot(141)
plt.title('Правило 1')
plt.axis(axis_value)
plt.plot(arr_x, func_1)
# Правило 2 - средне
plt.subplot(142)
plt.title('Правило 2')
plt.axis(axis_value)
plt.plot(arr_x, func_2)
# Правило 3 - средне
plt.subplot(143)
plt.title('Правило 3')
plt.axis(axis_value)
plt.plot(arr_x, func_3)
# Правило 4 - много
plt.subplot(144)
plt.title('Правило 4')
plt.axis(axis_value)
plt.plot(arr_x, func_4)
```

```
def max_term(f1, f2, f3, f4):
  res_arr = []
  for i in range(len(f1)):
    x, y, z, k = f1[i], f2[i], f3[i], f4[i]
    res_arr.append(max((x, y, z, k)))
  return res_arr
```

```
axis_value = [0, 11, -0.5, 1.5]
arr_x = np.linspace(0, 10, 1000)

res_func = max_term(func_1, func_2, func_3, func_4)
plt.axis(axis_value)
plt.plot(arr_x, res_func)
```

Найдем точку, где первый раз достигается максимум (где первый раз достигается 0.33333333333333337):

```
index = 0
for i, y in enumerate(res_func):
  if y == 0.33333333333333337:
    index = i
    break

result = arr_x[index]
print(result)
```

Исходя из полученного графика степени принадлежности выходного терма, можно сказать, что при объеме закупок 3.5, и качестве 0.55 понадобится не менее 1.001001001001001 цены (степень уверенности данного утверждения 0.33333333333333337).
