```
def calc_qual(price_v, buy_v, plot=False): 
    price_x = np.arange(many_pr['range'][0], many_pr['range'][1], 0.1)
    mask = np.isclose(price_x, 20, 1e-2)
    p_l = norm.pdf(price_x, sm_pr['mean'], sm_pr['width'])[mask][0]
    p_m = norm.pdf(price_x, mid_pr['mean'], mid_pr['width'])[mask][0]
    p_h = norm.pdf(price_x, many_pr['mean'], many_pr['width'])[mask][0]
    
    buy_x = np.arange(sm_buy['range'][0], sm_buy['range'][1], 0.1)
    mask = np.isclose(buy_x, 7)
    b_l = norm.pdf(buy_x, sm_buy['mean'], sm_buy['width'])[mask][0]
    b_m = norm.pdf(buy_x, mid_buy['mean'], mid_buy['width'])[mask][0]
    b_h = norm.pdf(buy_x, man_buy['mean'], man_buy['width'])[mask][0]

    qual_low_res = Rule2(p_m, b_l)
    qual_mid_res = np.maximum(Rule1(p_l, b_m), Rule6(p_h, b_l))
    qual_high_res = np.maximum(Rule3(p_h, b_h), np.maximum(Rule4(p_l, b_h), Rule5(p_m, b_h)))

    qual_x = np.arange(sm_quality['range'][0], sm_quality['range'][1], 0.1)
    acc = np.maximum(qual_mid_res, np.maximum(qual_low_res, qual_high_res))
    res = fuzz.centroid(qual_x, acc)

    if plot:
        plt.ylim(-0.1, 1.1)
        plt.plot(qual_x, acc)
        plt.scatter(res, 0, c='r')

    return res

def calc_points(price_list, buy_list):
    return [calc_qual(i, j) for i, j in zip(price_list, buy_list)]

def points_through_price(price_v, buy_v, number):
    res = [price_v]
    buy_history = [buy_v]
    i = 0
    while True:
        buy_history.append(buy_history[-1] + calc_qual(res[-1], buy_history[-1]))
        res.append(res[-1] - 5 * np.sin(buy_history[-1] * np.pi / 180))
        yield res[-1]
        i += 1
        if i == number:
            return
```
