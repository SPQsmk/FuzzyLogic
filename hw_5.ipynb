```
from Athene.reasoner.knowledgebase.axioms import *
from Athene.reasoner.common.constructors import *
import rdflib, rdflib.store


graph = default_world.as_rdflib_graph()


query1 =  "PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>"\
"PREFIX owl: <http://www.w3.org/2002/07/owl#>"\
"PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>"\
"PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>"\
"PREFIX a: <http://www.semanticweb.org/hw/ontologies/2020/10/untitled-ontology-4#>"\
"SELECT ?subject ?obj" \
"WHERE { " \
    "?subject rdfs:subClassOf ?obj"\
"}"


resultsList1 = graph.query(query1)
cars = list(onto.data_properties())[3]


response = []
classes = []
for item in resultsList1:
    subject = str(item['subject'].toPython())
    subject = re.sub(r'.*#', "", subject)
    classes.append(subject)
    response.append("Класс:" + subject)
print(response)


query2 = "PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>"\
"PREFIX owl: <http://www.w3.org/2002/07/owl#>"\
"PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>"\
"PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>"\
"PREFIX a: <http://www.semanticweb.org/spqsmk/ontologies/2020/10/untitled-ontology-2#>"\
   "SELECT *" \
        "WHERE{" \
            "?car rdf:type a:Кабриолет."\
            "?car a:Модель ?model."\
            "?car a:Содержит ?wheele."\
        "}"


resultsList = graph.query(query2)


response = []
cars = []
models = []
terms = []
for item in resultsList:
    print(item)
    car = str(item['car'].toPython())
    subject = re.sub(r'.*#', "", car)
    cars.append(subject)
    response.append("Автомобиль:" + subject)

    model = str(item['model'].toPython())
    subject = re.sub(r'.*#', "", model)
    models.append(subject)
    response.append("Модель:" + subject)

    term = str(item['wheele'].toPython())
    subject = re.sub(r'.*#', "", term)
    terms.append(subject)
    response.append("Запчасти:" + subject)
print(response)


from prettytable import PrettyTable  


x = PrettyTable()
print(cars)
x.add_column('Автомобиль', cars)
x.add_column('Модель', models)
x.add_column('Запчасти', terms)
print(x)
y = PrettyTable()
y.add_column('Класс', classes)
print(y)
```
