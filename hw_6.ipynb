1. Если Антон не убежден в чем-то, то он
убежден, что он в этом не убежден.
2. Если некоторое утверждение ложно, то Антон убежден, что он убежден в его истинности.
3. Если Антон в чем-то не убежден, он не
может быть убежден в противоположном.
4. Относительно любого утверждения Антон
убежден либо в его истинности, либо в истинности
противоположного утверждения.

>Допустим, Антон убежден, что если сдать лабораторную или курсовую, то он получит зачет. Следует ли из этого, что он убежден, что
если сдать лабораторную, то он получит зачет?

#### Запишем условия, используя модальный оператор B — `убежден`:

1. $\neg Bp \rightarrow B\neg Bp$
2. $\neg p \rightarrow BBp$
3. $\neg Bp\rightarrow \neg B\neg p$
4. $Bp \vee B\neg p$

#### l — утверждение о том, что если сдать лабораторную, то получит зачет.
#### k — утверждение о том, что если сдать курсовую, то получит зачет.
#### Нужно вывести: $B(l\vee k)\rightarrow Bl$
#### Вывод:

1. $\neg Bp \rightarrow B\neg Bp$ - аксиома 1
2. $\neg Bp\rightarrow \neg B\neg p$ - аксиома 3
3. $B\neg Bp\rightarrow \neg B\neg\neg Bp$ - подставим 2
4. $p\rightarrow \neg\neg p$ - тавтология
5. $B(p\rightarrow \neg\neg p)$ - обобщение 4
6. $B(Bp\rightarrow \neg\neg Bp)$ - подстановка 5
7. $B(p\rightarrow l)\rightarrow(Bp\rightarrow Bl)$ - аксиома К
8. $B(Bp\rightarrow \neg\neg Bp)\rightarrow(BBp\rightarrow B\neg\neg Bp)$ - подстановка 7
9. $BBp\rightarrow B\neg\neg Bp$ - modus ponens 6,8
10. $(p\rightarrow \neg l)\rightarrow((r\rightarrow l)\rightarrow(p\rightarrow \neg r))$ - тавтология
11. $(B\neg Bp\rightarrow \neg B\neg\neg Bp)\rightarrow((BBp\rightarrow B\neg\neg Bp)\rightarrow(B\neg Bp\rightarrow \neg BBp))$ - подстановка из 3, 9 в 10 
12. $(BBp\rightarrow B\neg\neg Bp)\rightarrow(B\neg Bp\rightarrow \neg BBp)$ - modus ponens 3, 11
13. $B\neg Bp\rightarrow \neg BBp$ - modus ponens 9, 12
14. $(p\rightarrow l)\rightarrow((l\rightarrow r)\rightarrow(p\rightarrow r))$ - тавтология
15. $(\neg Bp \rightarrow B\neg Bp)\rightarrow((B\neg Bp\rightarrow \neg BBp)\rightarrow(\neg Bp\rightarrow \neg BBp))$ - подстановка 1, 13
16. $(B\neg Bp\rightarrow \neg BBp)\rightarrow(\neg Bp\rightarrow \neg BBp)$ - modus ponens 1, 15
17. $\neg Bp\rightarrow \neg BBp$ - modus ponens 13, 16 
18. $\neg p \rightarrow BBp$ - аксиома 2
19. $(\neg Bp\rightarrow \neg BBp)\rightarrow((\neg p \rightarrow BBp)\rightarrow(\neg Bp\rightarrow \neg p))$ - подстановка 17, 18 в 10
20. $(\neg p \rightarrow BBp)\rightarrow(\neg Bp\rightarrow \neg p)$ - Modus ponens 17, 19
21. $\neg Bp\rightarrow \neg p$ - modus ponens 18, 20

#### Все, в чем не убежден Антон - ложно.
#### Дальше:

1. $\neg Bp\rightarrow \neg p$ - доказано
2. $\neg B(l\vee k)\rightarrow \neg (l\vee k)$ - из 1
3. $\neg B(l\vee k)\rightarrow \neg l$ - из 2
4. $\neg p\rightarrow \neg Bp$ - из 1
5. $Bp \vee B\neg p$ - аксиома 4
6. $\neg p \rightarrow B\neg p$ - из 4 и 5
7. $\neg l \rightarrow B\neg l$ - из 6
8. $\neg B(l\vee k)\rightarrow B\neg l$ - из 3 и 7

```
def check(graph, vertex, statement):
    return statement(graph, vertex)


def checkLinkedAnd(graph, vertex, statement):
    linked = graph.neighbors(vertex)
    for node in linked:
        if not statement(graph, node):
            return False
    return True


def checkLinkedOr(graph, vertex, statement):
    linked = graph.neighbors(vertex)
    for node in linked:
        if statement(graph, node):
            return True
    return False


def checkAllAnd(graph, statement):
    all = graph.nodes()
    for node in all:
        if not statement(graph, node):
            return False
    return True


def checkBoth(graph, vertexA, vertexB, statement):
    return statement(graph, vertexA) and statement(graph, vertexB)


def checkBothGraphs(graphA, graphB, vertexA, vertexB, statement):
    return statement(graphA, vertexA) and statement(graphA, vertexB)


def checkLaw(graph, statement):
    all = graph.nodes()
    for node in all:
        if not statement(graph, node):
            return False
    return True


def checkLawBoth(graph, vertexA, vertexB, statement):
    return statement(graph, vertexA) and statement(graph, vertexB)


def checkLawBothGraphs(graphA, graphB, vertexA, vertexB, statement):
    return statement(graphA, vertexA) and statement(graphA, vertexB)
```
