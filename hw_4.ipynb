$\quad$ Тогда, действительно $∃R.(A \sqcup B) ≡ ∃R.A \sqcup ∃R.B$

$\quad$ Остальные случаи доказываются аналогично

3. $∀R.⊥ ≡ ⊥ ≡ ∅, $ $∃R.\top ≡ \top ≡$ все первые элементы пар множества R
